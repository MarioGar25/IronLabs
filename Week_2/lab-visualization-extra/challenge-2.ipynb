#### Now in addition to the summary statistics, we also want to visualize the distribution of `Age`. Choose the appropriate plot to sow below.

```python
plot = sns.kdeplot(x=titanic["Age"], hue=titanic["Gender"])
plot.axvline(x=titanic["Age"].mean(), c="red", label="mean")
plot.axvline(x=titanic["Age"].dropna().median(), c="green", label="median")
plt.legend()
```

#### What additional information in the previous plot provide you about the feature `Age`? Comment.

In this graph it is clearer how the age is distributed, and when adding the gender we can also see the higher number of men and women. We also see that the majority of people are concentrated between 20 and 40 years of age, as we saw in the boxplot. In addition, the graph for men has a steeper curve, suggesting a greater number of people in the same age range, while the curve for women is smoother, implying a greater age distribution.
